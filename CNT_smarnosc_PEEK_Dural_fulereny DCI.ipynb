```python
def wieksze_niz (data):
    data.Time = data.Time.round(0)
    data = data[data.Current > 1]
    data = data.groupby("Time").sum()
    return data   

def concat(df_1,df_2,df_3):
    df_1=wieksze_niz (df_1)
    df_2=wieksze_niz (df_2)
    df_3=wieksze_niz (df_3)
    resultY=(pd.concat([df_1,df_2,df_3], sort=False, join='outer').groupby('Time')['Current']\
        .sum().reset_index())
    resultY.Current=resultY.Current/3
    return resultY

df_1Y=concat(df_1AA,df_1BB,df_1CC)
df_2Y=concat(df_2AA,df_2BB,df_2CC)
df_3Y=concat(df_3AA,df_3BB,df_3CC)
df_4Y=concat(df_4AA,df_4BB,df_4CC)
df_5Y=concat(df_5AA,df_5BB,df_5CC)
df_6Y=concat(df_6AA,df_6BB,df_6CC)
df_7Y=concat(df_7AA,df_7BB,df_7CC)
# df_41Y=concat(df_41T,df_41T2,df_41T3)
df_43Y=concat(df_43T,df_43T,df_43T)
```
